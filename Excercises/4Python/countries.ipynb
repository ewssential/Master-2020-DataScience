```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the countries.csv directly via URL import into your panda data frame!

url="https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/countries.csv"
countries=pd.read_csv(url)
countries
```

```python
#Display some basic information as rows, columns and some basic statistical info.

print(countries.shape)
print(countries.describe())
```

```python
#Show the last 4 rows of the data frame.

countries.tail(4)
```

```python
#Show all the row of countries who have the EURO

euro_countries = countries['Currency'] == 'EUR'
countries[euro_countries]
```

```python
#Show only name and Currency in a new data frame

new_data_farme = countries.filter(['Name','Currency'], axis=1).copy()
new_data_farme
```

```python
#Show only the rows/countries that have more than 2000 BIP (it is in Milliarden USD Bruttoinlandsprodukt)

rich_countries = countries['BIP'] > 2000
countries[rich_countries]
```

```python
#Select all countries where with inhabitants between 50 and 150 Mio
fifty_million = 50000000;
onehundred_fifty_million = 150000000;

medium_sized_countries = countries['People'].between(fifty_million,onehundred_fifty_million)
countries[medium_sized_countries]
```

```python
#Change BIP to Bip

countries.rename(columns={"BIP": "Bip"},inplace=True)
countries
```

```python
#Calculate the Bip sum

countries['Bip'].sum()
```

```python
#Calculate the average people of all countries

countries['People'].mean()
```

```python
#Sort by name alphabetically

countries.sort_values(by=['Name'], inplace=True)
countries
```

```python
#Create a new data frame from the original where the area is changed as follows: all countries with > 1000000 get BIG and <= 1000000 get SMALL in the cell replaced!
new_countries = countries.copy()
small_countries = new_countries['Area'] <= 1000000
new_countries['Area'] = np.where(small_countries, 'SMALL', 'BIG')

print(countries)
print(new_countries)
```
